# Describing Distributions

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from scipy import stats
```

```python
filename = 'Data/datadescription.csv'
df = pd.read_csv(filename)
df.head()
```

```python
vitD = df['VitD_levels']
mean_vitD = vitD.mean()
median_vitD = vitD.median()
std_vitD = vitD.std()
plus_one_std_vitD = mean_vitD + std_vitD
minus_one_std_vitD = mean_vitD - std_vitD
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(vitD, ax=ax, stat= 'probability', kde = True)
ax.axvline(mean_vitD, color = 'red', label = f'Mean = {mean_vitD:,.2f}',)
ax.axvline(median_vitD, ls='--', color = 'green', label = f'Median = {median_vitD:,.2f}')
ax.axvline(plus_one_std_vitD, color = 'black',label=f'+1 std = {plus_one_std_vitD:,.2f}')
ax.axvline(minus_one_std_vitD, color = 'black', label = f'-1 std = {minus_one_std_vitD:,.2f}')
ax.axvspan(plus_one_std_vitD, minus_one_std_vitD, color = 'yellow', zorder = 0)
ax.set_title(f'VitD_levels')
ax.legend();
```

```python
from scipy import stats
kurt_vit = stats.kurtosis(df['VitD_levels'], fisher = False)
kurt_vit
```

1. This is Continuous

```python
doc = df['Doc_visits']
mean_doc = doc.mean()
median_doc = doc.median()
std_doc = doc.std()
plus_one_std_doc = mean_doc + std_doc
minus_one_std_doc = mean_doc - std_doc
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(doc, ax=ax, stat= 'density', kde = True)
ax.axvline(mean_doc, color = 'red', label = f'Mean = {mean_doc:,.2f}',)
ax.axvline(median_doc, ls='--', color = 'green', label = f'Median = {median_doc:,.2f}')
ax.axvline(plus_one_std_doc, color = 'black',label=f'+1 std = {plus_one_std_doc:,.2f}')
ax.axvline(minus_one_std_doc, color = 'black', label = f'-1 std = {minus_one_std_doc:,.2f}')
ax.axvspan(plus_one_std_doc, minus_one_std_doc, color = 'yellow', zorder = 0)
ax.set_title(f'Doc Visits')
ax.legend();
```
